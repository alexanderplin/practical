```python
train = csv.reader((open('train.csv')))
next(train, None)

# dictionary with (user, [(artist1, numplays1), (artist2, numplays2) ...])
user_artist = {}

# row[0] = user id, row[1] = artist id
# this is fast; takes < 1 minute
print 'Begin:', str(datetime.now())
for i, row in enumerate(train):
	if user_artist.get(row[0]) == None:
		user_artist[row[0]] = {row[1]: row[2]}
	else:
		user_artist[row[0]][row[1]] = row[2]
	if i % 1000000 == 0:
		print 'Row', i

print 'End:', str(datetime.now())
```

```python
test = csv.reader((open('test.csv')))
next(test, None)

# dictionary with {user: {id: artist, id: artist}}
test_user_artist = {}

# row[0] = id, row[1] = user id, row[2] = artist id
print 'Begin:', str(datetime.now())
for i, row in enumerate(test):
	if test_user_artist.get(row[1]) == None:
		test_user_artist[row[1]] = {row[0]: row[2]}
	else:
		test_user_artist[row[1]][row[0]] = row[2]
	if i % 1000000 == 0:
		print 'Row', i

print 'End:', str(datetime.now())
```

```python
artists = pd.read_csv('artists.csv')
```

```python
# artists = artists.drop('name', axis=1)
artist_ids = artists['artist']
artist_ids.to_csv('artist-ids.csv', index=False)
```

```python
user_ids = users['user'].values

# takes about 30 seconds per 100 = 10 minutes
all_artists = []
for i, artist_id in enumerate(artist_ids):
    new_artist = []
    for user_id in user_ids:
        val = user_artist.get(user_id, 0)
        if val != 0:
            val = val.get(artist_id, 0)
        new_artist.append(val)
    all_artists.append(new_artist)
    if i % 100 == 0:
        print 'Iter', i
```

```python
users.columns
```

```python
for i, artist_list in enumerate(all_artists):
    users[artist_ids[i]] = pd.Series(artist_list, index=None, dtype=int)
    if i % 100 == 0:
        print 'Iter', i
```

```python
users.to_csv('one-hot-profiles-with-artists.csv', index=False)
```
